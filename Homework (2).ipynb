````python
!pip install -q datasketch

from datasketch import MinHash, MinHashLSH
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
````

````python
df_dedup = df_clean.sample(
    10000,
    random_state=42
).reset_index(drop=True)

df_dedup.shape
````

````python
# === Пример обоснования выбора модели ===

# 1. Точное совпадение
before = len(df_dedup)
df_exact = df_dedup.drop_duplicates(subset=["desc"])
after = len(df_exact)

print("Было:", before)
print("Стало:", after)
print("Удалено:", before-after)

# 2. Нечеткое сходство (MinHash или Levenshtein)
before = len(df_exact)

lsh = MinHashLSH(threshold=0.8, num_perm=128)
keep_idx = []

for i, text in enumerate(df_exact["desc"]):
    m = MinHash(num_perm=128)
    for word in text.split():
        m.update(word.encode("utf8"))

    if not lsh.query(m):
        lsh.insert(str(i), m)
        keep_idx.append(i)

df_minhash = df_exact.iloc[keep_idx].reset_index(drop=True)

after = len(df_minhash)

print("Было:", before)
print("Стало:", after)
print("Удалено:", before - after)


# 3. Семантическое сходство (sentence-transformers)
model_name = "intfloat/multilingual-e5-base"
model = SentenceTransformer(model_name)

texts = df_minhash["desc"].tolist()

embeddings = model.encode(
    texts,
    batch_size=64,
    show_progress_bar=True,
    convert_to_numpy=True
)

cos_sim = cosine_similarity(embeddings)
np.fill_diagonal(cos_sim, 0)

threshold = 0.95
to_drop = set()

for i in range(len(cos_sim)):
    if i in to_drop:
        continue
    similar_indices = np.where(cos_sim[i] > threshold)[0]
    for j in similar_indices:
        to_drop.add(j)

before = len(df_minhash)
df_semantic = (
    df_minhash
    .drop(index=list(to_drop))
    .reset_index(drop=True)
)

after = len(df_semantic)

print("Было:", before)
print("Стало:", after)
print("Удалено:", before - after)
````

**Обоснование выбора LLM для разметки:**

Я выбрал модель Qwen/Qwen2.5-0.5B-Instruct для локального инференса через библиотеку transformers. Она отлично справляется с задачами следования инструкциям и форматированием вывода, при этом имеет очень маленький размер.
Альтернативы вроде Llama-3.2-1B требуют больше памяти, а API-сервисы требуют регистрации и платных токенов.

````python
import torch
import json
import re

from transformers import AutoModelForCausalLM, AutoTokenizer
from sklearn.metrics import accuracy_score, f1_score, classification_report

model_id = "Qwen/Qwen2.5-0.5B-Instruct"

tokenizer = AutoTokenizer.from_pretrained(model_id)

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    device_map="auto"
)

model.eval()

golden_set = df_dedup.sample(
    100,
    random_state=42
).reset_index(drop=True)

system_prompt = """Ты классификатор медицинских жалоб.Выбери одного врача.
Ответ строго JSON:
{
"specialization": "..."
}

Варианты:
Гастроэнтеролог,
Гинеколог,
Дерматолог,
Невролог,
Отоларинголог,
Проктолог,
Стоматолог,
Терапевт,
Травматолог,
Уролог,
Хирург
"""

def extract_specialization(text):
    try:
        data = json.loads(text)
        answer = data.get("specialization")

        if answer in classes:
            return answer
    except:
        pass

    for c in classes:
        if c.lower() in text.lower():
            return c
    return "UNKNOWN"

def predict_zero_shot(text):
    messages = [
        {
            "role": "system",
            "content": system_prompt
        },
        {
            "role": "user",
            "content": text
        }
    ]

    inputs = tokenizer.apply_chat_template(
    messages,
    tokenize=True,
    add_generation_prompt=True,
    return_tensors="pt",
    return_dict=True
    )

    inputs = inputs.to(model.device)
    input_ids = inputs["input_ids"]
    input_length = input_ids.shape[1]

    with torch.no_grad():
        outputs = model.generate(
        input_ids=input_ids,
        max_new_tokens=30,
        do_sample=False
    )

    response = tokenizer.decode(
        outputs[0][input_length:],
        skip_special_tokens=True
    ).strip()

    return extract_specialization(response)

print("Размечаем Golden Set...")
golden_set["pred_zero_shot"] = golden_set["desc"].apply(
    predict_zero_shot
)
golden_set["pred_zero_shot"] = golden_set["pred_zero_shot"].replace(
    "UNKNOWN",
    "Терапевт"
)
print(
    golden_set["pred_zero_shot"].value_counts()
)

accuracy = accuracy_score(
    golden_set["spec10"],
    golden_set["pred_zero_shot"]
)

f1 = f1_score(
    golden_set["spec10"],
    golden_set["pred_zero_shot"],
    average="macro",
    zero_division=0
)

print("Accuracy:", accuracy)
print("F1-macro:", f1)
print(
    classification_report(
        golden_set["spec10"],
        golden_set["pred_zero_shot"],
        zero_division=0
    )
)
````
